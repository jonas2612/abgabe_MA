```python
import pandas as pd
import seaborn as sns
```

```python
ds = ['exvito', 'gastrulation', 'invito']
genes = ['10k', 'all']
metric = ['ari', 'ASW', 'batch_ASW', 'cLISI', 'graph_connectivity', 'iLISI', 'iso_ASW', 'nmi', 'pcr', 'iso_F1']
```

```python
data = []
for d in ds:
    for m in metric:
        tmp = pd.read_csv(f'{d}_{m}_unintegrated.csv')
        tmp['integrated'] = 'unintegrated'
        tmp['n_genes'] = '10k_genes'
        if m =='ari':
            tmp['metric'] = 'ari_unintegrated'
        elif m=='iso_F1':
            tmp['metric'] = 'iso_F1_unintegrated'
        data.append(tmp)
        tmp = pd.read_csv(f'{d}_{m}_unintegrated.csv')
        tmp['integrated'] = 'unintegrated'
        tmp['n_genes'] = 'all_genes'
        if m =='ari':
            tmp['metric'] = 'ari_unintegrated'
        elif m=='iso_F1':
            tmp['metric'] = 'iso_F1_unintegrated'
        data.append(tmp)
        for g in genes:
            tmp = pd.read_csv(f'{d}_{g}_genes_{m}_integrated.csv')
            tmp['integrated'] = 'integrated'
            data.append(tmp)
            tmp = pd.read_csv(f'{d}_{g}_genes_traject_conservation.csv')
            tmp['integrated'] = 'integrated'
            data.append(tmp)
data = pd.concat(data)
data
```

```python
invitro = data[data.dataset=='invito']
exvitro = data[data.dataset=='exvito']
gastrulation = data[data.dataset=='gastrulation']
```
