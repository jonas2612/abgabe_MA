```python
import scanpy as sc
```

```python
gast = sc.read_h5ad('gastrulation_ds.h5ad')
sc.pl.umap(gast, color='celltype', save='_gastrulation_celltype.png')
sc.pl.umap(gast, color='day', save='_gastrulation_day.png')
sc.pl.umap(gast, color='batch', save='_gastrulation_batch.png')
```
