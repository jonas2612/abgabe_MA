```python
import scanpy as sc
```

```python
gastrulation = sc.read_h5ad('/home/icb/jonas.flor/gastrulation_atlas/data/gastrulation_IDs.h5ad')
invitro = sc.read_h5ad('/home/icb/jonas.flor/gastrulation_atlas/data/invitro.h5ad')
exvitro = sc.read_h5ad('/home/icb/jonas.flor/gastrulation_atlas/data/exvitro.h5ad')
```
